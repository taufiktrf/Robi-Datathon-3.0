```python
## importing libraries
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', 30)
```

```python
## read the given csv files

boxes = pd.read_csv('boxes.csv')
prob_3 = pd.read_csv("problem 2.csv")
submission_2 = pd.read_csv("sample submission 2.csv")
purchase = pd.read_csv('purchase.csv')
```

```python
#filter the MAGIC_KEY that are in the problem_2 file
filtered_purchase_03 = purchase[purchase["MAGIC_KEY"].isin(prob_3["MAGIC_KEY"])]
```

```python
#change the date to pandas datetime
filtered_purchase_03['PURCHASE_DATE'] = pd.to_datetime(filtered_purchase_03['PURCHASE_DATE'], dayfirst=True, errors='coerce')
```

```python
#make seperate columns for date, month and year
filtered_purchase_03['PURCHASE_MONTH'] = filtered_purchase_03['PURCHASE_DATE'].dt.month
filtered_purchase_03['PURCHASE_YEAR'] = filtered_purchase_03['PURCHASE_DATE'].dt.year
filtered_purchase_03['PURCHASE_DAY'] = filtered_purchase_03['PURCHASE_DATE'].dt.day
```

```python
#merge the filtered_purchase dataframe with boxs.csv on box_id
data_03 =  pd.merge(filtered_purchase_03,boxes, on='BOX_ID', how='left')
```

```python
## checking the columns 
data_03
```

```python
data_03=data_03.dropna()
```

```python
#calculate total purchase 
total_purchase = data_03.groupby('MAGIC_KEY').size()
total_purchase.name = 'TOTAL_PURCHASE'
```

```python
#make a new column named TOTAL_PURCHASE
data_03 = data_03.merge(total_purchase, on='MAGIC_KEY', how='left')
```

```python
#sort the dataframe on TOTAL_PURCHASE, MAGIC_KEY, PURCHASE_COUNT
data_03 = data_03.sort_values(by=['TOTAL_PURCHASE', 'MAGIC_KEY', 'PURCHASE_DATE'], ascending=[False, True, True])
data_03
```

```python
#cheking magic_key for certain range
ids = data_03[(data_03['TOTAL_PURCHASE'] > 5) & (data_03['TOTAL_PURCHASE'] < 15)]['MAGIC_KEY'].values
```

```python
#calculating average gap between purchasing meat
avg_gap = {}
for key, grp in data_03[data_03['MAGIC_KEY'].isin(ids)].groupby('MAGIC_KEY'):
    avg_gap[key] = grp[grp['MEAT'] > 0]['PURCHASE_DATE'].diff().mean()
```

```python
avg_gap
```

```python
# those who buy meat and milk both are prone to buying meat at a certain gap.
# we are checking magic_key if they purchased meat and milk both. If true calculate the average_gap_between
# the_meat_purchase. Now check for the time between the last meat purchase, if greater than the average_gap_between_the
# meat_purchase then select the last purchased box_id with meat as next purchase.

# but those MAGIC_KEY buying only milk or only milk are prone to buy the last purchaed BOX_ID. Select the last
# purchased box_id as their next purchase

next_purchase = {}
for key, grp in data_03.groupby('MAGIC_KEY'):
    grp['m'] = (grp['MEAT'] > 0).values
    if 2 < grp['m'].sum() < grp['TOTAL_PURCHASE'].values[0]:
        avg_gap = grp[grp['MEAT'] > 0]['PURCHASE_DATE'].diff().mean().days
        last_index = grp[grp['m'] == True].tail(1).index.values[0]
        last_date, box_id = grp.loc[last_index, ['PURCHASE_DATE', 'BOX_ID']].values
        time_delta = (pd.Timestamp('2019-03-01') - last_date).days
        if avg_gap <= time_delta < 45:
            next_purchase[key] = box_id
        else:          
            last_index = grp[grp['m'] == False].tail(1).index.values[0]
            next_purchase[key] = grp.loc[last_index, 'BOX_ID']
    else:
        next_purchase[key] = grp.loc[grp.index[-1], 'BOX_ID']
```

```python
## renaming in the submission format
next_purchase_df = pd.DataFrame({'MAGIC_KEY':next_purchase.keys(), 'BOX_ID':next_purchase.values()})
next_purchase_df
```

```python
sub = pd.merge(prob_3, next_purchase_df, on='MAGIC_KEY', how='left')
sub.head(50)
```

```python
sub.to_csv('submission.csv', index=False)
```
